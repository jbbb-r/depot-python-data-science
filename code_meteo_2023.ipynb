```python
# Lecture avec autodétection du séparateur
df23bis = pd.read_csv(
    "data/2023_comptage-velo-donnees-compteurs.csv",
    sep=None,
    engine='python'
)

# Garder uniquement les colonnes que tu veux
colsToKeep = [
    "Identifiant du point de comptage",
    "Nom du point de comptage",
    "Comptage horaire",
    "Date et heure de comptage",
    "Lien vers photo du point de comptage",
    "Coordonnées géographiques"
]

# Renommer les colonnes (elles sont actuellement en 0,1,2,...)
df23bis.columns = colsToKeep

# Filtrer les lignes de header parasites éventuelles
df23bis = df23bis[df23bis["Nom du point de comptage"] != "Nom du point de comptage"]


df23bis.head()
```

```python
# Supprimer deux colonnes
df23bis = df23bis.drop(columns=[
    "Identifiant du point de comptage",
    "Lien vers photo du point de comptage"
])

# Renommer une colonne
df23bis = df23bis.rename(columns={
    "Nom du point de comptage": "Nom du site de comptage"
})

df23bis.head()
```

```python
# Séparer latitude et longitude
df23bis[['latitude', 'longitude']] = df23bis['Coordonnées géographiques'].str.split(',', n=1, expand=True)

df23bis["Date et heure de comptage"] = pd.to_datetime(
    df23bis["Date et heure de comptage"], 
    utc=True,    # rend les dates tz‑aware (UTC)
    errors='coerce'
)
```

```python
cols = ["Nom du site de comptage", "Coordonnées géographiques", "latitude", "longitude"]
df23bis_sites = df23bis[cols].drop_duplicates(subset=["Nom du site de comptage"])
print(df23bis_sites.shape)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df23bis_sites
```

```python
import geopandas as gpd

# 1. Copie du DataFrame de base
df = df23bis_sites.copy()

# 2. Nettoyage des coordonnées
df['latitude']  = pd.to_numeric(df['latitude'],  errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

df = df.dropna(subset=['latitude', 'longitude'])

# 3. Création d’un GeoDataFrame à partir du DataFrame nettoyé
gdf_sites = gpd.GeoDataFrame(
    df,
    geometry = gpd.points_from_xy(df.longitude, df.latitude),
    crs="EPSG:4326"
)

# 4. Chargement et projection des arrondissements
url_geojson = "https://opendata.paris.fr/explore/dataset/arrondissements/download/?format=geojson&timezone=Europe/Berlin&lang=fr"
gdf_arr = gpd.read_file(url_geojson)
gdf_arr = gdf_arr.to_crs(epsg=4326)

# 5. Spatial join — points dans les polygones
gdf_joined = gpd.sjoin(gdf_sites, gdf_arr, how="left", predicate="within")

# 6. Sélection & renommage des colonnes d’intérêt
df23bis_final = gdf_joined[
    ["Nom du site de comptage", "Coordonnées géographiques", "latitude", "longitude", "l_ar"]
].rename(columns={"l_ar": "arrondissement"})

df23bis_final
```

```python
import pandas as pd

def list_to_semicolon_str(lst):
    return "; ".join(str(v) for v in lst if pd.notna(v))

# Regroupement par arrondissement et concaténation des coordonnées
tableau_arr = (
    df23bis_final
    .groupby("arrondissement", dropna=False)
    .agg({
        "Coordonnées géographiques": lambda x: list_to_semicolon_str(x),
        "latitude": lambda x: list_to_semicolon_str(x),
        "longitude": lambda x: list_to_semicolon_str(x)
    })
    .reset_index()
)

# Trier les arrondissements par numéro — si le nom contient un nombre
# On extrait le premier nombre trouvé dans la chaîne (ex “12e”, “8”, etc.)
tableau_arr['arr_num'] = (
    tableau_arr['arrondissement']
    .str.extract(r'(\d+)', expand=False)  # capture les chiffres
    .astype(float, errors='ignore')       # convertit en float si possible
)

# Trier par ce numéro, en mettant les NaN (arrondissements sans numéro numérique) à la fin
tableau_arr = tableau_arr.sort_values(
    by='arr_num', na_position='last'
).drop(columns='arr_num').reset_index(drop=True)

# Optionnel : afficher tout, mais attention si le tableau est très long
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

tableau_arr
```

```python
import requests
import pandas as pd
import calendar
import os
import time

# -------------------------------
# PARAMÈTRES
# -------------------------------
ANNEE = 2023
TIMEZONE = "Europe/Paris"
VARIABLE_METEO = "apparent_temperature"

DOSSIER_DATA = "data"
DOSSIER_MENSUEL = os.path.join(DOSSIER_DATA, "mensuel")
DOSSIER_HORAIRE = os.path.join(DOSSIER_DATA, "horaire")

os.makedirs(DOSSIER_MENSUEL, exist_ok=True)
os.makedirs(DOSSIER_HORAIRE, exist_ok=True)

# -------------------------------
# FONCTIONS UTILES
# -------------------------------
def dernier_jour_du_mois(annee, mois):
    return calendar.monthrange(annee, mois)[1]

def liste_mois(annee):
    return [f"{annee}-{m:02d}" for m in range(1, 13)]

# -------------------------------
# LISTES DE STOCKAGE
# -------------------------------
results_mensuel = []
results_horaire = []

mois = liste_mois(ANNEE)

# -------------------------------
# BOUCLE PRINCIPALE
# -------------------------------
for arr, df_arr in df23bis_final.groupby("arrondissement"):
    ligne_arr = {"arrondissement": arr}

    for periode in mois:
        an, m = map(int, periode.split("-"))
        start = f"{an}-{m:02d}-01"
        end = f"{an}-{m:02d}-{dernier_jour_du_mois(an, m)}"

        valeurs_mois = []

        for _, row in df_arr.iterrows():
            lat = float(row["latitude"])
            lon = float(row["longitude"])

            url = (
                "https://archive-api.open-meteo.com/v1/archive?"
                f"latitude={lat}&longitude={lon}"
                f"&start_date={start}&end_date={end}"
                f"&hourly={VARIABLE_METEO}"
                f"&timezone={TIMEZONE}"
            )

            try:
                resp = requests.get(url, timeout=30)
                resp.raise_for_status()
                data = resp.json()

                if "hourly" in data and VARIABLE_METEO in data["hourly"]:
                    heures = data["hourly"]["time"]
                    valeurs = data["hourly"][VARIABLE_METEO]

                    # 🔹 Stockage horaire
                    for h, v in zip(heures, valeurs):
                        results_horaire.append({
                            "annee": ANNEE,
                            "arrondissement": arr,
                            "latitude": lat,
                            "longitude": lon,
                            "datetime": h,
                            VARIABLE_METEO: v
                        })

                    # 🔹 Stockage pour moyenne mensuelle
                    valeurs_mois.extend(valeurs)

                time.sleep(0.2)  # anti-saturation API

            except Exception as e:
                print(f"❌ Erreur {arr} {periode} ({lat},{lon}) : {e}")

        # Moyenne mensuelle arrondissement
        ligne_arr[periode] = (
            sum(valeurs_mois) / len(valeurs_mois)
            if valeurs_mois else None
        )

    results_mensuel.append(ligne_arr)

# -------------------------------
# DATAFRAME MENSUEL
# -------------------------------
df_mensuel = pd.DataFrame(results_mensuel)

df_mensuel["arr_num"] = (
    df_mensuel["arrondissement"]
    .str.extract(r"(\d+)")
    .astype(float)
)

df_mensuel = (
    df_mensuel
    .sort_values("arr_num", na_position="last")
    .drop(columns="arr_num")
    .reset_index(drop=True)
)

# -------------------------------
# DATAFRAME HORAIRE
# -------------------------------
df_horaire = pd.DataFrame(results_horaire)

# -------------------------------
# SAUVEGARDES
# -------------------------------
csv_mensuel = os.path.join(
    DOSSIER_MENSUEL,
    f"{ANNEE}_temperature_mensuelle_moyenne_par_arrondissement.csv"
)

csv_horaire = os.path.join(
    DOSSIER_HORAIRE,
    f"{ANNEE}_temperature_horaire_par_coordonnees.csv.gz"
)

df_mensuel.to_csv(csv_mensuel, index=False)
df_horaire.to_csv(csv_horaire, index=False, compression="gzip")

print("✅ Fichiers générés avec succès :")
print(f" - {csv_mensuel}")
print(f" - {csv_horaire}")
```
